```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = r"C:\Users\user\Desktop\study\ml_read\housing.csv"
data = pd.read_csv(path)

prices = data['MEDV']
features = data.drop('MEDV', axis=1)
```

```python
features.head()
```

```python
data.describe()
```

```python
min_price = np.min(prices)
max_price = np.max(prices)
mean_price = np.mean(prices)
median_price = np.median(prices)
std_price = np.std(prices)

print("Minimum price: ${:,.2f}".format(min_price))
print("Maximum price: ${:,.2f}".format(max_price))
print("Mean price: ${:,.2f}".format(mean_price))
print("Median price: ${:,.2f}".format(median_price))
print("Standard deviation of prices: ${:,.2f}".format(std_price))
```

```python
plt.figure(figsize=(12, 5))
plt.hist(data['MEDV'], bins=50, color='skyblue', edgecolor='black')
plt.title('Распределение цен жилья (MEDV)', fontsize=15)
plt.xlabel('Цена жилья (MEDV)', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(data['MEDV'], bins=50, kde=True, color='skyblue', edgecolor='black')
plt.title('Распределение цен жилья (MEDV)', fontsize=15)
plt.xlabel('Цена жилья (MEDV)', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.show()
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(16,5))

for i in range(len(axs)):
    axs[i].set_ylim([0, 1200000])
    
_ = sns.regplot(x=features['RM'], y=prices, ax=axs[0])
_ = sns.regplot(x=features['LSTAT'], y=prices, ax=axs[1], color='orange')
_ = sns.regplot(x=features['PTRATIO'], y=prices, ax=axs[2], color='green')

plt.show()
```
